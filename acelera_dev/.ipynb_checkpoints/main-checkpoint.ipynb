```python
import sklearn
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
regr = LinearRegression()
```

```python
df_train = pd.read_csv('train.csv')
df_train = df_train.dropna(0)
```

```python
x = df_train[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC']]
y = df_train.NU_NOTA_MT
assert(len(x) == len(y))
```

```python
x.columns
```

```python
plt.figure()
plt.xlabel('NU_NOTA_COMP1')
plt.ylabel('Nota de matemática')
plt.title('Nota de matemática x NU_NOTA_CN')
colors = ['b', 'g']
plt.plot(x.NU_NOTA_CN, y)
plt.axis([0, 800, 0, 800])
plt.grid()
plt.show()
```

```python
# Relacionamento de ciência da natureza vs matemática

colors = ['b', 'g']
plt.scatter(x.NU_NOTA_CN, y, c=colors)
plt.xlabel('Nota de ciência da natureza')
plt.ylabel('Nota de Matemática')
plt.title('Relação matematica com ciências da natureza')
plt.show()
```

```python
# Relacionamento de ciência da natureza vs matemática
colors = ['b', 'g']
plt.scatter(x.NU_NOTA_CH, y, c=colors)
plt.xlabel('Nota de ciência da Humanas')
plt.ylabel('Nota de Matemática')
plt.title('Relação matematica com ciência da Humanas')
plt.show()
```

```python
# Relacionamento de ciência da natureza vs matemática
colors = ['b', 'g']
plt.scatter(x.NU_NOTA_LC, y, c=colors)
plt.xlabel('Nota linguagem de código')
plt.ylabel('Nota de Matemática')
plt.title('Relação matematica com linguagem de código')
plt.show()
```

```python
# coêficientes
regr_cn = LinearRegression()
regr_cn.fit(x, y)
```

```python
regr_cn.predict(x)
```

```python
colors = ['b', 'g']
plt.scatter(y, regr_cn.predict(x), c=colors)
plt.xlabel('Notas Originais ')
plt.ylabel('Nota  prevista')
plt.title('Notas Originais vs Nota  prevista')
plt.show()
```

```python
print(f'Coêficiente {regr_cn.intercept_}')
print(f'n coeficiente {len(regr_cn.coef_)}')
print(f'Mean SQUARE ERROR (MSE){np.mean((regr_cn.predict(x- y)**2))}')
```
