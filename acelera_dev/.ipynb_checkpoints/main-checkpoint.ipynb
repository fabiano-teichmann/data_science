```python
import sklearn
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
regr = LinearRegression()
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
df_train.dropna(0)
```

```python
x = df_train[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC']]
y = df_train.NU_NOTA_MT
assert(len(x) == len(y))
```

```python
x = x.dropna(0)
```

```python
y = y.dropna(0)
```

```python
# Relacionamento de ciência da natureza vs matemática
plt.scatter(x.NU_NOTA_CN, y)
plt.xlabel('Nota de ciência da natureza')
plt.ylabel('Nota de Matemática')
plt.title('Relação matematica com ciências da natureza')
plt.show()
```

```python
# Relacionamento de ciência da natureza vs matemática
plt.scatter(x.NU_NOTA_CH, y)
plt.xlabel('Nota de ciência da natureza')
plt.ylabel('Nota de Matemática')
plt.title('Relação matematica com ciências da natureza')
plt.show()
```

```python
# coêficientes
regr_cn = LinearRegression()
regr_cn.fit([x.NU_NOTA_CN], [y])
```

```python
print(f'Coêficiente {regr_cn.intercept_}')
print(f'n coeficiente {len(regr_cn.coef_)}')
```
